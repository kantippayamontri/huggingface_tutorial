* text generation task
    * more general task that may not involve a conversational context.
    * broad task -> just give prompt, topic or context
* chat completion task
    * the model is aware of the conversation context (the user's input) and responds accordingly

```python
from huggingface_hub import InferenceClient
```

```python
# InferenceClient = helper use to talk to HG servers
client = InferenceClient(
    model="mistralai/Mistral-7B-Instruct-v0.2", # any public model
    api_key="" # your hf api key
)
```

```python
response = client.chat_completion(
    messages=[{
        "role": "user",
        "content": "what's machine learning?"
    }]
)
print(response)
```

```python
print(type(response))
for c in response.choices:
    print(c.message.content)
```
